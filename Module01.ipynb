```python
#Import Dataset
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Maryland_Players_Drafted.csv')
print(df.head())  
```

```python
#Clean Up + Grouping

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df.dropna(subset=['Year'])
draft_counts = df.groupby('Year').size().reset_index(name='Number of Players Drafted')

draft_counts = draft_counts.sort_values('Year')
```

```python
# All Years Graph
plt.figure(figsize=(14, 7))
plt.bar(draft_counts['Year'], draft_counts['Number of Players Drafted'], color='darkgreen')
plt.xlabel('Year')
plt.ylabel('Number of Players Drafted')
plt.title('NFL Players Drafted Each Year (All Years)')
plt.xticks(draft_counts['Year'][::5], rotation=45)  
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Clean Up + Grouping for -20 Year Graph

df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
recent_df = df[df['Year'] >= (df['Year'].max() - 19)]
draft_counts = recent_df.groupby('Year').size().reset_index(name='Number of Players Drafted')

draft_counts = draft_counts.sort_values('Year')
```

```python
#Graph Last 20 Years
plt.figure(figsize=(12, 6))
plt.bar(draft_counts['Year'], draft_counts['Number of Players Drafted'], color='steelblue')
plt.xlabel('Year')
plt.ylabel('Number of Players Drafted')
plt.title('NFL Players Drafted Each Year (Last 20 Years)')
plt.xticks(draft_counts['Year'], rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```
